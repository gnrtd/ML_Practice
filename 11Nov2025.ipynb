```python
from pandas.tseries.holiday import USFederalHolidayCalendar #any country
from pandas.tseries.offsets import CustomBusinessDay #from that calendar
```

```python
usb = CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
rng = pd.date_range(start='7/1/2017', end='7/31/2017', freq=usb)
```

```python
rng
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday
#create my own calendar with birthday holidays
```

```python
#Blueprint
class myBirthdayCalendar(AbstractHolidayCalendar):
  rules = [
      Holiday('Olesia Bday', month=8, day=24)
  ]
```

```python
myc = CustomBusinessDay(calendar=myBirthdayCalendar())
```

```python
rng = pd.date_range(start='8/1/2017', end='8/30/2017', freq=myc)
```
