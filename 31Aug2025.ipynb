```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
income  = np.random.exponential(scale=50000, size= 100).astype(int)
```

```python
income
```

```python
df_exp = pd.DataFrame({
    'ID': range(1,101),
    'Income': income
})
```

```python
df_exp
```

```python
import seaborn as sns
sns.histplot(df_exp['Income'])
```

```python
df_exp['log_income'] = np.log(df_exp['Income'])
```

```python
df_exp
```

```python
sns.histplot(df_exp['log_income'])
```

```python
num1 =  float(input('Enter first number: '))
num2 = float(input('Enter second number: '))
operation = input('Choose operation (+, -, *, /): ')
# perform calculation
if operation == '+':
  result = num1 + num2
elif operation == '-':
  result = num1 - num2
elif operation == '*':
  result = num1 * num2
elif operation == '/':
  result = num1 / num2
else:
  result = 'Invalid operation. Try again'
# display result
print('Result: ', result)
```
